```python
def load_labeled_videos(directory, frames_per_video, frame_interval=5):
    video_data = []

    for class_name in os.listdir(directory):
        class_path = os.path.join(directory, class_name)

        if os.path.isdir(class_path):
            class_frames = [] 

            for filename in os.listdir(class_path):
                frames = []  

                if filename.endswith((".mp4", ".avi")):
                    video_path = os.path.join(class_path, filename)

                 
                    cap = cv2.VideoCapture(video_path)

        
                    if not cap.isOpened():
                        print(f"Error opening video file: {video_path}")
                        continue

                    
                    frame_counter = 0

                    while True:
                        ret, frame = cap.read()
                        if not ret:
                            break

                        if frame_counter % frame_interval == 0:
                            frames.append(frame)

                        frame_counter += 1
                        if len(frames) >= frames_per_video:
                            break

                    
                    cap.release()

                    if len(frames) < frames_per_video:
                        print(f"Warning: Insufficient frames in {video_path}. Skipping.")

                class_frames.extend(frames)  

            video_data.append({'frames': class_frames, 'label': class_name})

    return video_data
```

```python
directory_path = 'Cdata'
labeled_videos = load_labeled_videos(directory_path, frames_per_video=20)
```

```python
def helper(num):
    if num ==0:
        return "workers"
    if num == 1:
        return "customers"
    if num == 2:
        return "empty"
    if num == 3:
        return "far away interaction"
    if num == 4:
        return "Register interaction"
    
```

```python
print(labeled_videos[0]['label'])
```

```python
frames = []
labels = []
class_to_label = {}  # Mapping class names to integer labels

for class_data in labeled_videos:
    frames.extend(class_data['frames'])
    
    # Assuming class_data['label'] contains the class name
    class_name = class_data['label']
    
    # Assign a unique integer label to each class
    if class_name not in class_to_label:
        class_to_label[class_name] = len(class_to_label)
    
    # Use the integer label as the class label
    label = class_to_label[class_name]
    
    # Extend labels for each frame
    labels.extend([label] * len(class_data['frames']))
```

```python
print(len(frames))
print(labels)
```

```python
data_augmentation_hor = tf.keras.Sequential([ tf.keras.layers.RandomFlip(mode="horizontal") ])
```

```python
data_augmentation_ver = tf.keras.Sequential([ tf.keras.layers.RandomFlip(mode="vertical") ])
```

```python
common_size = (224, 224)  
frames_resized = [cv2.resize(video, common_size) for video in frames]


X_train, X_test, y_train, y_test = train_test_split(frames_resized, labels, test_size=0.2, random_state=42)


y_train = np.array(y_train)
y_test = np.array(y_test)


X_train_resized = np.array(X_train)
X_test_resized = np.array(X_test)


X_train_resized = X_train_resized / 255.0
X_test_resized = X_test_resized / 255.0


print("X_train_resized shape:", X_train_resized.shape)
print("X_test_resized shape:", X_test_resized.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
print("X_train_resized shape:", X_train_resized.shape)
print("X_test_resized shape:", X_test_resized.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
inputshape=(224, 224, 3)
model = Sequential([
    
    layers.Input(shape=inputshape,name='input1'),
    layers.Conv2D(32, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(3, 3)),
    data_augmentation_hor,
 
   
    layers.Conv2D(32, kernel_size=(3, 3), activation='relu'),

    layers.MaxPooling2D(pool_size=(3, 3)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.7),
    layers.Dense(5, activation='softmax')
                        ])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_accuracy', patience = 4, mode = 'max', verbose = 1,restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, min_lr=1e-5)
history = model.fit(X_train_resized,y_train, epochs=30, validation_data=(X_test_resized,y_test),callbacks=[early_stopping,reduce_lr])
```

```python
def plotAccuracy(history):

    try:
        plt.plot(history.history['accuracy'])
        plt.plot(history.history['val_accuracy'])
    except KeyError:
        plt.plot(history.history['acc'])
        plt.plot(history.history['val_acc'])
    plt.title('Accuracy vs. epochs')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Training', 'Validation'], loc='lower right')
    plt.show() 
plotAccuracy(history)
```

```python
def plotLoss(history):
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Loss vs. epochs')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Training', 'Validation'], loc='upper right')
    plt.show() 
plotLoss(history)
```

```python
def videochecker(curr_video, modeli):
    cap = cv2.VideoCapture(curr_video)
    worker=0
    customer=0
    register_interaction =0
    far_away_interaction =0
    empty = 0
    count =0
    TT =cap.get(cv2.CAP_PROP_FRAME_COUNT)
    print(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    while True:
        print(count)
        ret, frame = cap.read()
        if  ret:
            count +=20;
            cap.set(cv2.CAP_PROP_POS_FRAMES, count)
        else:
            cap.release()
            break

        img2 = cv2.resize(frame, (224, 224))
        i_array = np.array(img2)
        i_array = np.expand_dims(i_array, axis=0) 
        predictions_single = model.predict(i_array)
        predicted_class_index = np.argmax(predictions_single)
        print(helper(predicted_class_index))
        curpre = helper(predicted_class_index)
    
        if curpre== "workers":
            worker+=20

        if curpre == "customers":
            customer+=20
        if curpre == "empty":
            empty +=20
        if curpre =="far away interaction":    
            far_away_interaction +=20
        if curpre == "Register interaction":
            register_interaction +=20
  
    print("Video consiste of")
    print("Workers %",(worker/count)*100)
    print("Customers %",(customer/count)*100)
    print("empty %",(empty/count)*100)
    print("register interaction %",(register_interaction/count)*100)
    print("far_away_interaction %",(far_away_interaction/count)*100)
    
```
